```python
import turtle
tl = turtle.Turtle()
tl.shape("turtle")
hight = 200
wide = 300
tl.penup()
tl.goto(-120,-70)
tl.pendown()
tl.forward(wide)
tl.left(90)
tl.forward(hight)
tl.left(90)
tl.forward(wide)
tl.fillcolor('green')
tl.begin_fill()
tl.left(90)
tl.forward(hight)
tl.left(90)
tl.penup()
tl.forward(hight/2)
tl.left(90)
tl.pendown()
tl.forward(hight)
tl.end_fill()
tl.fillcolor('red')
tl.begin_fill()
tl.right(180)
tl.penup()
tl.forward(hight)
tl.left(90)
tl.forward(hight/2)
tl.pendown()
tl.left(90)
tl.forward(hight)
tl.end_fill()
tl.fillcolor('yellow')
tl.begin_fill()
tl.penup()
tl.right(90)
tl.forward(hight/2)
tl.right(90)
tl.forward(hight)
tl.right(90)
tl.forward(hight/2)
tl.end_fill()
tl.right(90)
tl.forward(95)
tl.left(90)
tl.forward(25)
tl.pendown()
tl.pencolor('yellow')
tl.fillcolor('yellow')
tl.begin_fill()
tl.forward(50)
tl.left(36)
tl.backward(50)
tl.right(144)
tl.backward(50)
tl.left(36)
tl.forward(50)
tl.left(36)
tl.backward(50)
tl.end_fill()
turtle.done()
```
